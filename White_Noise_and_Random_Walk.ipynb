# Random Walk and White Noise Ljung Box Test

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from statsmodels.stats.diagnostic import acorr_ljungbox
```

# Synthetic Data Generation

```python
n = 1000;

#White Noise
white_noise = np.random.normal(0,1,n)

#Random Walk
random_shocks = np.random.normal(0,1,n)
random_walk = np.cumsum(random_shocks)
```

# Data Plotting

```python
plt.figure(figsize=(12,6))

plt.subplot(211)
plt.plot(white_noise)
plt.title('White Noise')

plt.subplot(212)
plt.plot(random_walk)
plt.title(f'Random Walk')

plt.tight_layout()
```

# Ljung Box Test

```python
print("Ljung Box Test for White Noise:")
lb_test_white_noise = acorr_ljungbox(white_noise,lags=[10],return_df=True)
print(lb_test_white_noise)

#Since p-value is greater than 0.05, it is White Noise

print("Ljung Box Test for Random Walk:")
lb_test_white_noise = acorr_ljungbox(random_walk,lags=[10],return_df=True)
print(lb_test_white_noise)

#Since p-value is lesser than 0.05, it is Random Walk
```

# ACF and PACF plot

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plt.figure(figsize=(12,6))
#d=2

#PACF plot
plt.subplot(121)
plot_pacf(white_noise,ax=plt.gca(), lags=50, method='ywm')
plt.title('Partial Autocorrelation Factor (PACF) White Noise')
#p=7

#ACF plot
plt.subplot(122)
plot_acf(white_noise,ax=plt.gca(), lags=50)
plt.title('Autocorrelation Factor (ACF) White Noise')
#q=1
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plt.figure(figsize=(12,6))
#d=2

#PACF plot
plt.subplot(121)
plot_pacf(random_walk,ax=plt.gca(), lags=50, method='ywm')
plt.title('Partial Autocorrelation Factor (PACF) Random Walk')
#p=7

#ACF plot
plt.subplot(122)
plot_acf(random_walk,ax=plt.gca(), lags=50)
plt.title('Autocorrelation Factor (ACF) Random Walk')
#q=1
```
